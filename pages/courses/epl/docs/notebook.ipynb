```python
# one can customize and make the enum callable
# one can hence wrap behaviour. 
# ok this is virtuous and useless ;-)

class MySort(Enum):
    ASCENDING = 1
    DESCENDING = 2

    # it makes the members (the enum entries) callable
    def __call__(self, values):
        # here self is the enumEntry
        # evaluate what self is now
        isDESCENDING = self is MySort.DESCENDING
        return sorted(values, reverse=isDESCENDING)

l = [3,5,1,0.1,10]
print(MySort.ASCENDING(l))
print(MySort.DESCENDING(l))
```

```python
# IntFlags

from enum import IntFlag

# power of two flags
class Role(IntFlag):
    OWNER = 8
    POWER_USER = 4
    USER = 2
    SUPERVISOR = 1
    # or operator 'add' to the enum the flag
    ADMIN = OWNER | POWER_USER | USER | SUPERVISOR

john_roles = Role.USER | Role.SUPERVISOR

if Role.USER in john_roles:
    print("John, you're a user")

if Role.SUPERVISOR in john_roles:
    print("John, you're a supervisor")

print(Role.OWNER in Role.ADMIN)
print(Role.SUPERVISOR in Role.ADMIN)

# | operator is summing
print(Role.ADMIN.value)

# if 'flag' instead of 'intflag' you cannot do this
print(Role.ADMIN+1)
```

```python

# functions can be recursed. 
# "Iteration is human, recursion is divine" (Peter Deutsch, inventor of ghostscripts and many other things)


# 1,1,2,3
# fibonacci(4) = 3 = 2+1 = fibonacci(3)+fibonacci(2)

def fibonacci_recursive(n):    
    if n <= 1:
        return n
    else:                        
        one_step = fibonacci_recursive(n-1)            
        two_steps = fibonacci_recursive(n-2)
        return  one_step+two_steps         

#               (3)
#              F(4)
#          ↙          ↘
#         (2)         (1)
#        F(3)         F(2)
#       ↙   ↘        ↙   ↘
#      (1)   (1)    (1)  (0)
#     F(2)  F(1)   F(1)  F(0)
#    ↙   ↘
#  (1)   (0)
#  F(1)  F(0)

fibonacci_recursive(4)
```

```python
# operators overloading 

class Point3d():
    
    description = 'I am a point in 3d'

    def __init__(self,x=0,y=0,z=0):
        self.x = x
        self.y = y
        self.z = z
    
    def makeString(self):
        return f'my position is {self.x},{self.y},{self.z}'
    
    def printPosition(self):
        print(self.makeString())

    def __str__(self):        
        return Point3d.description+' and '+self.makeString()

    def customFunction(self):
        print('I am point3d custom function')
            
    @staticmethod
    def printDescription():
        print(Point3d.description)        

    def __add__(self,anotherPoint):
        newPoint = Point3d()
        newPoint.x = self.x+anotherPoint.x
        newPoint.y = self.y+anotherPoint.y
        newPoint.z = self.z+anotherPoint.z
        return newPoint
        # there is no type checking, hence you can do anything! don't do it
        #return newPoint.makeString()

    def __eq__(self,anotherPoint):
        newPoint = Point3d()
        if (newPoint.x == self.x+anotherPoint.x and 
            newPoint.y == self.y+anotherPoint.y and
            newPoint.z == self.z+anotherPoint.z):   
            return True
        else:
            return False
        

point1 = Point3d(0,1,2)
point2 = Point3d(1,2,3)

point3 = point1+point2

print(point3)

print(point1==point2)
```

```python
## inheritance and methods ovverride

class Atom(Point3d):
    
    def __init__(self,x,y,z,element,charge=0):
        super().__init__(x,y,z)
        self.element = element
        self.charge = charge        

    def __str__(self):
        return f'I am an atom, element = {self.element} and '+super().makeString()

    # method ovverride
    def customFunction(self):
        print('I am an atom custom function')

    # how to recover an overridden method
    def customFunctionPoint(self):
        super().customFunction()

    def giveMeSuper(self):
        return super()

atom = Atom(0,1,2,'H')        
print(atom)
atom.customFunction()
atom.customFunctionPoint()

# this doest not work
#atom.super().customFunction()

# ... but I can cheat to recover the mother class function
atom.giveMeSuper().customFunction()
```

```python
## composition example, iterators

class Molecule():
    
    def __init__(self,name: str):
        self.name = name
        self.atomsList = []          
        self._current_index = 0
    
    def addAtom(self,at: Atom): 
        self.atomsList.append(at)

    # it returns the object on which the next() function has to be called
    # when iterating
    def __iter__(self):
        self._current_index = 0
        return self
        #return iter(self.atomsList)
    
    # it has to return the 'next' object in our object
    def __next__(self):        
        if self._current_index < len(self.atomsList):
            val = self.atomsList[self._current_index]        
            self._current_index += 1
            return val
        else:             
             raise StopIteration   
        
    def getAtom(self,i):
        return self.atomsList[i]

    def __len__(self):
        return len(self.atomsList)    

    def __str__(self):
        fullStr = ''
        # i am using the fact i can iterate
        for a in self:
            # i am using the fact the Atom __str__ function is implemented
            fullStr += str(a)+'\n'
        return fullStr
    

at1 = Atom(0,1,2,'C')  
at2 = Atom(3,4,5,'O')  
at3 = Atom(6,7,8,'O')  

mol = Molecule('CO2')
mol.addAtom(at1)
mol.addAtom(at2)
mol.addAtom(at3)

print('Manual iteration')
for i in range(0,len(mol)): 
    print(mol.getAtom(i))

print('Through iterators')
for atom in mol:
    print(atom)

print('Molecule __str__')
print(mol)
```

```python

# a function can work as a class decorator
def counter_decorator(classVar):  
    # global counter  
    count = 0

    class DecoratorWrapper:
        # instance counter
        #count = 0
        # intercepts the constructor args
        def __init__(self,*args):
            print('just wrapped')  
            # repropagate them
            self.wrapped = classVar(*args)                 

        # this propagates the '.' get element operator
        # try to comment  
        
        def __getattr__(self, name):
            nonlocal count            
            #self.count += 1
            count += 1
            #print('retrieving count %d'%self.count)            
            print('retrieving count %d'%count)                        
            return getattr(self.wrapped,name)
                      

    return DecoratorWrapper

@counter_decorator
class myClass:
    ciao = 'ciao'
    def __init__(self):
        print('init')

mc = myClass()   
mc2 = myClass()   
mc.ciao     
mc.ciao     
mc2.ciao     
mc2.ciao     
```

```python
# decorators with arguments

def switcherDecorator(switchMe):
    # this captures the function pointer
    def functionCatcher(F):
        # this capture the arguments
        def argsCatcher(*args,**kwargs):
            #print('switch %d'%switchMe)
            # if switch, call the passed function not the orginal one
            if (switchMe):                  
                return args[0]()
            # if does not switch, call the the usual function
            else:
                return F()        
        return argsCatcher
    return functionCatcher


@switcherDecorator(True)
def firstFunction(F=None):
    print('first function')

def secondFunction(F=None):
    print('second function')

firstFunction(secondFunction)
```

```python

# can arbitrarily modify the behaviour of a code ...
# analogous to a PRELOAD low level linux behaviour
def hackDecorator(switchMe):
    def functionCatcher(F):
        def argsCatcher(*args,**kwargs):
            #print('switch %d'%switchMe)
            if (switchMe):                  
                return "license not expired"                
            else:
                return F()        
        return argsCatcher
    return functionCatcher

# hack the behaviour of a function
@hackDecorator(True)
def checkLicense():
    print('license expired')
       
checkLicense()        
```

```python

# decorator function for building the singleton pattern

def makeSingleton(myClass):    
    instance = None
    def onCall(*args,**kwargs):
        nonlocal instance
        # if the instance does not exist, create it
        if (instance == None):
            instance = myClass(*args)
        # if the instance already exists, do nothing
        else:
            print('I am a singleton...sorry')
            pass
        return instance
    return onCall


@makeSingleton
class myTargetClass():    
    def __init__(self,name):        
        self.name = name


me = myTargetClass('Mario')
print(me.name)

# you can do that, but now is a singleton, you cannot do that anymore
me2 = myTargetClass('Luigi')

print(me2.name)
```

```python
# decorator for limited number of instances

instances = list()
maxInstancesNumber = 10

def makeSingleton(myClass):    
    global instances     
    def onCall(*args,**kwargs):        
        lastInstance = None
        if (len(instances)<maxInstancesNumber):
            lastInstance = myClass(*args,**kwargs)
            instances.append(lastInstance)                
        else:
            print('Cannot allocate more, returns the last available')
            return  instances[maxInstancesNumber-1]
        return lastInstance
    return onCall

@makeSingleton
class myTargetClass():    
    def __init__(self,name):
        self.myName = name        


for i in range(0,maxInstancesNumber+1):
    me = myTargetClass(str(i))
    print(me.myName)
```

```python

## EXERCISES 

def insertion_sort(arr):

    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]  # Current element to be sorted

        # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1            
        arr[j + 1] = key

    return arr

my_array = [4, 2, 7, 1, 5,10,0,-3]
sorted_array = insertion_sort(my_array)
print(sorted_array)

```

```python
import numpy as np
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt


def gradient_descent(cost_function, initial_params, learning_rate, num_iterations,toll=1e-3):
    params = initial_params.copy()
    history = []  # List to store the cost at each iteration

    for i in range(num_iterations):
        cost = cost_function(params)
        history.append(cost)

        gradient = compute_gradient(cost_function, params)
        params -= learning_rate * gradient        
        
        if (np.abs(np.max(gradient))<toll):            
            return params,history
        
    return params, history

# finite differences, you can try it analytical
def compute_gradient(cost_function, params, epsilon=1e-8):
    num_params = len(params)
    gradient = np.zeros(num_params)

    for i in range(num_params):
        delta = np.zeros(num_params)
        delta[i] = epsilon

        cost_plus = cost_function(params + delta)
        cost_minus = cost_function(params - delta)

        gradient[i] = (cost_plus - cost_minus) / (2 * epsilon)

    return gradient

def cost_function(params):
    x, y = params
    return (x - 3.0) ** 2.0 + (y + 2.0) ** 2.0

initial_params = np.array([5.0, 3.0])
learning_rate = 0.03
num_iterations = 100

optimized_params, cost_history = gradient_descent(cost_function, initial_params, learning_rate, num_iterations)
print("Optimized parameters:", optimized_params)
print("Final cost:", cost_history[-1])
print("Num iterations:", len(cost_history))


# build an animation

plt.plot(cost_history)
plt.show()

fig, ax = plt.subplots()
ax.set_xlim(0, +5)
ax.set_ylim(-4, +5)
line, = ax.plot([], [], 'r', marker='o')

def animate(i):    
    #print(initial_params)
    optimized_params, _ = gradient_descent(cost_function, initial_params, learning_rate, i+1)      
    line.set_data(optimized_params[0], optimized_params[1])
    return line,

animation = FuncAnimation(fig, animate, frames=num_iterations, interval=100)

from IPython.display import HTML
HTML(animation.to_jshtml())

```
